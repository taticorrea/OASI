```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits
from astropy.time import Time
import statistics as std
plt.style.use('classic')
%matplotlib inline
```

```python
it = ['2019-01-01T21:00:00']                             #Inicio das observaçoes
itime = Time(it, format = 'isot', scale = 'utc' )
```

```python
dados_40s = pd.read_csv('seeing_40s.csv',header = 0, names = ['DATE-OBS', 
                                                              'FILTER',
                                                              'EXP-TIME',
                                                              'AIRMASS',
                                                              'FWHM',
                                                              'XBINNING',
                                                              'SEEING',
                                                              'IMAGE-NAME'])  
dados_40s
```

```python
seeing_15_40s = dados_40s[(dados_40s.SEEING < 1.5)]
```

```python
lista_tempos_40s = Time(list(seeing_15_40s['DATE-OBS']),format='isot',scale='utc')

dt = []
t_40s = np.empty(len(seeing_15_40s['DATE-OBS']))

for i in range(len(lista_tempos_40s)):
    dt.append(lista_tempos_40s[i] - itime)
    t_40s[i] = (dt[i].value)/30
```

```python
dados_30s = pd.read_csv('seeing_30s.csv',header = 0, names = ['DATE-OBS', 
                                                              'FILTER',
                                                              'EXP-TIME',
                                                              'AIRMASS',
                                                              'FWHM',
                                                              'XBINNING',
                                                              'SEEING',
                                                              'IMAGE-NAME'])  
dados_30s
```

```python
seeing_15_30s = dados_30s[(dados_30s.SEEING < 1.5)]
```

```python
lista_tempos_30s = Time(list(seeing_15_30s['DATE-OBS']),format='isot',scale='utc')

dt = []
t_30s = np.empty(len(seeing_15_30s['DATE-OBS']))

for i in range(len(lista_tempos_30s)):
    dt.append(lista_tempos_30s[i] - itime)
    t_30s[i] = (dt[i].value)/30
```

```python
plt.figure(figsize=(9,7))
plt.scatter(t_30s,seeing_15_30s['SEEING'], label = 'Filtro R\nTempo de exposição = 30s', color = 'darkorange', alpha = 1,lw = 0.8,edgecolor = 'black')
plt.scatter(t_40s,seeing_15_40s['SEEING'], label = 'Filtro R\nTempo de exposição = 40s', color = 'darkblue', alpha = 1,lw = 0.8,edgecolor = 'black')
plt.xticks(np.arange(13),('Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez'), rotation = 45, size = 15)
plt.xlim(0,12)
plt.yticks(np.arange(1,2,0.5))

plt.xlabel('Tempo de observação',size = 15)
plt.ylabel('Seeing (")',size = 15)
plt.legend(prop={'size': 10},frameon = True,loc = 'upper right', fancybox = True, framealpha = 1, shadow = True, borderpad = 0.5 )
plt.grid(axis = 'x')
plt.savefig('seeingxtempo.png')
plt.show()
```
