```python
%matplotlib inline
import statistics as std
import pandas as pd                                 
import numpy as np                          #Importando os módulos necessários
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.time import Time
from astropy.io.fits import HDUList
```

```python
#imagens usadas no psfmeasure = 1,51,101,151,201,251,301,351,401,501,551,565

image= '1998NU_40s_b2x2-001_R.fit'
name = []                                                        #Lista com o nome das imagens

for i in range(0,600,50):       
    if i < 10:                                                      #Para imagens 001 ate 009 
        number_raw = image[16:19].replace(image[18],str(int(image[18])+i))   
        name.append('1998NU_40s_b2x2-' + str(number_raw) + '_R.fit' )
    elif i >= 10 and i < 100:                                       #Para imagens 010 ate 099 
        image = '1998NU_40s_b2x2-011_R.fit'
        number_raw = image[16:18].replace(image[18],str(int(image[18])+i))
        name.append('1998NU_40s_b2x2-' + str(number_raw) + '_R.fit' )
for i in range(1,600,50):                                           #Para imagens >100 
    if i >= 100:
        image = '1998NU_40s_b2x2-100_R.fit'
        number_raw = image[17:18].replace(image[18],str(int(image[18])+i))
        name.append('1998NU_40s_b2x2-' + str(number_raw) + '_R.fit' )
        
del name[9]                                                      #451_R.fit ruim, troquei por 565_R.fit
name.append('1998NU_40s_b2x2-565_R.fit')
```

```python
date_obs = []                                                #Lista com as datas

for i in name:                                        #Iterando dentro da lista com os nomes
    with fits.open(i) as hdul:                           #Abrindo cada elemento da lista imagens
        for hdu in hdul:                                 #Extraindo o header de cada imagem
            date_obs.append(hdul[0].header['DATE-OBS'])

filtro = []                                                #Lista com as datas

for i in name:                                        #Iterando dentro da lista com os nomes
    with fits.open(i) as hdul:                           #Abrindo cada elemento da lista imagens
        for hdu in hdul:                                 #Extraindo o header de cada imagem
            filtro.append(hdul[0].header['FILTER'])

exp_time = []                                                #Lista com as datas

for i in name:                                        #Iterando dentro da lista com os nomes
    with fits.open(i) as hdul:                           #Abrindo cada elemento da lista imagens
        for hdu in hdul:                                 #Extraindo o header de cada imagem
            exp_time.append(hdul[0].header['EXPTIME'])

airmass = []                                                #Lista com as datas

for i in name:                                        #Iterando dentro da lista com os nomes
    with fits.open(i) as hdul:                           #Abrindo cada elemento da lista imagens
        for hdu in hdul:                                 #Extraindo o header de cada imagem
            airmass.append(hdul[0].header['AIRMASS'])

file = open('logfile','r')                       #Abrindo o arquivo com os dados
fwhm = []                                       #Lista vazia para salvar os valores do fwhm

for indice, linha in enumerate(file):            #Loop para ler as linhas apartir da 480 e pegar o fwhm
    if indice  > 1:
        if linha[:47] == '  Average full width at half maximum (FWHM) of ': 
            fwhm.append(float(linha[47:]))
```

```python
d = {'DATE-OBS':date_obs,'EXP-TIME':exp_time,'AIRMASS':airmass,'FILTER':filtro,'FWHM':fwhm,'NAME':name}
df = pd.DataFrame(data=d)
df = df.sort_values(by='DATE-OBS')
df.to_csv('2019-01-08_dados.csv')
df
```
