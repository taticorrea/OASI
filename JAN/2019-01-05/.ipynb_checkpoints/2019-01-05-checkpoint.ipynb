<h2>2002LJ3</h2>

```python
%matplotlib inline
import statistics as std
import pandas as pd                                 
import numpy as np                                         #Importando os m칩dulos necess치rios
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.time import Time
```

<h3>Raw data</h3>

```python
rfile = open('JAN/2019-01-05/2002LJ3_2019-01-05_dados_brutos.csv','r')                #Abrindo o arquivo com os dados
rdF = pd.read_csv(rfile)                                   #Lendo e salvando os dados na vari치vel dF

rdados = list(rdF['DATE-OBS'])                             #Transformando a coluna DATE-OBS em uma lista
rfwhm = list(rdF['FWHM'])                                  #Transformando a coluna FWHM em uma lista

rtimes = Time(rdados, format = 'isot', scale = 'utc')      #Instanciando a classe Time com os dados

it = ['2019-01-05T21:00:00']
itime = Time(it, format = 'isot', scale = 'utc' )

rfwhmm = np.empty(len(rfwhm))                              #Iniciando array vazio com tamanho da coluna fwhm
rdt = []                                                   #Inicializando a lista com os pontos
rt = np.empty(len(rdados))                                 #Inicializando array para plotar

for i in range(len(rfwhm)):
    rfwhmm[i] = rfwhm[i].replace(',','.')                  #Tirando o ',' e colocar '.' para decimal
    rdt.append(rtimes[i] - itime)
    rt[i] = (rdt[i].value)*24                              #Calculando os valores de dt em HORAS

rFWHM = list(rfwhmm)                                       #Lista FWHM
```

<h3>Calibrated data</h3>

```python
cfile = open('JAN/2019-01-05/2002LJ3_2019-01-05_dados_calibrados.csv','r')                #Abrindo o arquivo com os dados
cdF = pd.read_csv(cfile)                                   #Lendo e salvando os dados na vari치vel dF

cdados = list(cdF['DATE-OBS'])                             #Transformando a coluna DATE-OBS em uma lista
cfwhm = list(cdF['FWHM'])                                  #Transformando a coluna FWHM em uma lista

ctimes = Time(cdados, format = 'isot', scale = 'utc')      #Instanciando a classe Time com os dados

it = ['2019-01-05T21:00:00']
itime = Time(it, format = 'isot', scale = 'utc' )

cfwhmm = np.empty(len(cfwhm))                              #Iniciando array vazio com tamanho da coluna fwhm
cdt = []                                                   #Inicializando a lista com os pontos
ct = np.empty(len(cdados))                                 #Inicializando array para plotar

for i in range(len(cfwhm)):
    cfwhmm[i] = cfwhm[i].replace(',','.')                  #Tirando o ',' e colocar '.' para decimal
    cdt.append(ctimes[i] - itime)
    ct[i] = (cdt[i].value)*24                              #Calculando os valores de dt em HORAS

cFWHM = list(cfwhmm)                                       #Convertendo fwhm em lista
```

```python
plt.figure(figsize=(9,7))
plt.plot(ct, cFWHM, color = 'blue', label = 'calibrated data')              #Calibrated
plt.plot(rt, rFWHM, color = 'green', label = 'raw data')                    #Raw
plt.scatter(ct, cFWHM, color = 'blue')                                      #Calibrated
plt.scatter(rt, rFWHM, color = 'green')                                     #Raw
plt.yticks(np.arange(2, 5, 1)) 
plt.xticks(np.arange(0,7,1))                                                #Raw
plt.title('2019-01-05 Filtro R 80s')
plt.ylabel('Fwhm')
plt.xlabel('Time (hours)')
plt.savefig('fwhmxtime.png')
plt.legend()
plt.show()
```
